Now, let's solve the problem in hand. We have the below lists of items and categories, say.

````python
# lists of items and categories
items = ['eggs', 'apples', 'mobile', 'sofa']
categories = ['Food', 'Food', 'Electronics', 'Furniture']
````

````python
# Function to get the category of a given item name
def get_category(item_name):
    # List of items in the store
    items = ['eggs', 'apples', 'mobile', 'sofa']

    # Corresponding categories for each item
    categories = ['Food', 'Food', 'Electronics', 'Furniture']

    try:
        # Find the index of the given item in the items list
        index = items.index(item_name)

        # Return a list containing the item name and its category
        return [item_name, categories[index]]

    except ValueError:
        # If the item is not found in the list, return a message saying so
        return [item_name, "Item not found"]
````

Now that support can identify the category of a product, the inventory team wants to go the other way around:

We’ll create a function that does exactly that.

````python
# Function to get all items belonging to a specific category
def get_items_by_category(category_name):
    # List of items in the store
    items = ['eggs', 'apples', 'mobile', 'sofa']

    # Corresponding categories for each item
    categories = ['Food', 'Food', 'Electronics', 'Furniture']

    # Empty list to store items that belong to the given category
    items_in_category = []

    # Loop through all items by their index
    for i in range(len(items)):
        # Check if the category of the current item matches the given category
        if categories[i] == category_name:
            # If yes, add the item to our result list
            items_in_category.append(items[i])

    # Return the list of items that matched the category
    return items_in_category
````

###1.8 Understanding Class and Objects

As the data complexity increases, it becomes necessary to structure it in an organized way. This is where Object-Oriented Programming (OOP) comes into play. By using classes and objects, we can create models that accurately represent the real-world entities within WOW Superstore

#### 1.8.1 Defining the Order Class

A class is a blueprint for creating objects. It defines the attributes and methods that the objects created from the class will have.

- Attributes represent the data associated with the object.

- In our example, the Order class will represent a single order record and store all the relevant order details.

- Methods are functions defined within the class that specify the behaviors or actions the object can perform.

- In the Order class, methods will allow us to perform operations on an order, such as summarizing its details, applying discounts, or checking if it is profitable.

````python
class Order:
    # Constructor Method (__init__)
    # The constructor is a special method that runs automatically
    # when a new object of this class is created.
    # It initializes the object's attributes with provided values.
    def __init__(self, order_id, ship_mode, segment, city, state, region,
                 category, sub_category, sales, quantity, discount, profit):

        self.order_id = order_id
        self.ship_mode = ship_mode
        self.segment = segment
        self.city = city
        self.state = state
        self.region = region
        self.category = category
        self.sub_category = sub_category
        self.sales = sales
        self.quantity = quantity
        self.discount = discount
        self.profit = profit

    def get_summary(self):
        return f"Order {self.order_id} ({self.category}) - Sales: ${self.sales}, Profit: ${self.profit}"

    def apply_discount(self, extra_discount):
        # Increase discount
        self.discount += extra_discount

        # Assume 70% of sales = cost
        cost_price = self.sales * 0.7
        selling_price = self.sales * (1 - self.discount)

        # New profit
        self.profit = selling_price - cost_price

    def is_profitable(self):
        return self.profit > 0
````

####1.8.6 Flexible Order Handling


WOW Superstore receives **domestic and international orders**.  

- Domestic orders are simple.  
- International orders have extra shipping fees.  

We want a system that can **show order summaries in different ways** and **calculate profits correctly**.  


Base class (Domestic Order)

````python
class Order:
    """
    Represents a regular (domestic) order in WOW Superstore.
    """
    def __init__(self, order_id, category, sales, profit):
        self.order_id = order_id  # Unique order ID
        self.category = category  # Product category
        self.sales = sales        # Total sales amount
        self.profit = profit      # Profit from this order

    def get_summary(self, show_sales=True, show_profit=True):
        """
        Method overloading example: Same method name,
        different outputs depending on arguments.
        - show_sales=False -> hide sales
        - show_profit=False -> hide profit
        """
        parts = [f"Order {self.order_id} ({self.category})"]
        if show_sales:
            parts.append(f"Sales: ${self.sales}")
        if show_profit:
            parts.append(f"Profit: ${self.profit}")
        return " - ".join(parts)
````

Inherited class (International Order)

````python
class InternationalOrder(Order):
    """
    Represents an international order, which has extra shipping fees.
    """
    def __init__(self, order_id, category, sales, profit, shipping_fee):
        super().__init__(order_id, category, sales, profit)  # Inherit base attributes
        self.shipping_fee = shipping_fee  # Additional shipping cost

    def get_summary(self, show_sales=True, show_profit=True):
        """
        Method overriding example: Customize behavior for international orders.
        Profit is reduced by the shipping fee.
        """
        parts = [f"International Order {self.order_id} ({self.category})"]
        if show_sales:
            parts.append(f"Sales: ${self.sales}")
        if show_profit:
            parts.append(f"Profit after shipping: ${self.profit - self.shipping_fee}")
        return " - ".join(parts)
````

````python
# Domestic order example
order1 = Order("CA-2023-0001", "Furniture", 300, 50)

# International order example
intl_order1 = InternationalOrder("IN-2023-0001", "Electronics", 500, 80, 25)
````

````python
# Method Overloading: flexible summary for domestic order
print(order1.get_summary())                    # Full summary: Sales + Profit
print(order1.get_summary(show_profit=False))   # Only sales (hide profit)

# Method Overriding: international order adjusts profit
print(intl_order1.get_summary())               # Shows profit after shipping fee
````
