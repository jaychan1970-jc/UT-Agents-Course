```python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama",  # required, but ignored by Ollama
)

response = client.chat.completions.create(
    model="phi3",
    messages=[
        {"role": "system", "content": "You are a helpful local AI assistant."},
        {"role": "user", "content": "Say hello and tell me in one sentence what model you are."}
    ],
    temperature=0.2,
)

print(response.choices[0].message.content)
```

```python
tasks = [
    "Summarize this in 2 bullets: Python is a popular programming language used for data analysis, automation, web development, and AI.",
    "Write a Python function to check whether a number is prime.",
    "Explain list comprehensions in Python like I am a beginner."
]

for i, prompt in enumerate(tasks, 1):
    response = client.chat.completions.create(
        model="phi3",
        messages=[
            {"role": "system", "content": "You are a concise coding tutor."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.3,
    )
    print(f"\n--- Task {i} ---")
    print(response.choices[0].message.content)
```

```python
response = client.responses.create(
    model="phi3",
    input="Give me a short study plan for learning Python in 2 weeks."
)

print(response.output_text)
```
