```python
import os
from openai import OpenAI

# Option 1: put your key directly here (simplest for testing; not for production)
#OPENROUTER_API_KEY = "YOUR_OPENROUTER_API_KEY_HERE"
OPENROUTER_API_KEY = "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"

# Option 2 (better): store in an env var instead and read it:
# OPENROUTER_API_KEY = os.environ.get("OPENROUTER_API_KEY")

if not OPENROUTER_API_KEY:
    raise ValueError("Set OPENROUTER_API_KEY before running this cell.")

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=OPENROUTER_API_KEY,
    # OpenRouter recommends sending a referer + app name in headers
    default_headers={
        "HTTP-Referer": "http://localhost",   # or your site / repo URL
        "X-Title": "VSCode Jupyter Gemma Test"
    },
)
```

```python
response = client.chat.completions.create(
    model="google/gemma-4-31b-it",  # Gemma 4 31B Instruct on OpenRouter
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Hello Gemma 4 31B, briefly introduce yourself."}
    ],
    temperature=0.3,
)

print(response.choices[0].message.content)
```
